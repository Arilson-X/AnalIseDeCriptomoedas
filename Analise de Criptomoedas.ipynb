## Analise da cripto Aave
* Calculo do multiplo de Mayer
* Verificação entre o multiplo de Mayer e o valor em que fechou cada dia

```python
df_Aave = pd.read_csv("Base de Dados\coin_Aave.csv", delimiter = ",")
display(df_Aave)
```

### Cálculo do Multiplo de Mayer para Aave

```python
df_Aave_ultimos_200_dias = df_Aave.loc[df_Aave["SNo"] > 75]
m_Mayer_Aave = df_Aave_ultimos_200_dias["Close"].mean()
print(m_Mayer_Aave)
df_Aave_ultimos_200_dias = df_Aave_ultimos_200_dias.assign(Mayer = m_Mayer_Aave)
display(df_Aave_ultimos_200_dias[["Close","Mayer"]])
```
